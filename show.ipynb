```python
#导入相应模块
import re
import os 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
path = 'D:/split_dataset/'
files = os.listdir(path)
files_csv = list(filter(lambda x: x[-4:]=='.csv' , files))
#函数re.compile将正则表达式（以字符串书写的）转换为模式对象
df= []
num_filter = re.compile(r'\d+')
for file in  files_csv:
    tmp = pd.read_csv(path + file,encoding='gbk',low_memory=False,index_col=0)
    print(tmp.shape[1]-1)
    df.append(tmp)
final_df=df[0]
n=len(files_csv)
for i in range(n-1):
   final_df=pd.merge(final_df,df[i+1],on='entname',how='outer') 
title=final_df.columns.values.tolist()
```

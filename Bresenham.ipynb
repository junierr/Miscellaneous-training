```python
#直线
import matplotlib.pyplot as plt
xx=[]
yy=[]
x1=1
y1=1
x2=10
y2=6
x=x1
y=y1
d=(x2-x1)-2*(y2-y1)
delta1=(y2-y1)*2-2*(x2-x1)
delta2=(y2-y1)*2
while x<=x2:
    xx.append(x)
    yy.append(y)
    x+=1
    if(d>=0):
        d+=delta1
        y+=1
    else:
        d+=delta2
        
plt.plot(xx, yy,  color='r',markerfacecolor='blue',marker='o')  
for a, b in zip(xx, yy):  
    plt.text(a, b, (a,b),ha='center', va='bottom', fontsize=10)  
 
plt.legend()  
plt.show()
```

```python
#圆
import matplotlib.pyplot as plt
def bresenhamcircle(R):
    xx1=[]
    yy1=[]
    xx2=[]
    yy2=[]
    x=0
    y=R
    p=3-2*R
    while x<=y:
        xx1.append(x)
        yy1.append(y)
        if p>=0:
            p+=4*(x-y)+10
            y-=1
        else:
            p+=4*x+6
        x+=1
    for i in xx1:
        xx2.append(i*-1)
    for i in yy1:
        yy2.append(i*-1)
    
    plt.plot(xx1, yy1,  color='r',markerfacecolor='blue',marker='o')
    plt.plot(xx1, yy2,  color='r',markerfacecolor='blue',marker='o')
    plt.plot(yy1, xx2,  color='r',markerfacecolor='blue',marker='o')
    plt.plot(yy2, xx2,  color='r',markerfacecolor='blue',marker='o')
    plt.plot(xx2, yy2,  color='r',markerfacecolor='blue',marker='o')
    plt.plot(xx2, yy1,  color='r',markerfacecolor='blue',marker='o')
    plt.plot(yy2, xx1,  color='r',markerfacecolor='blue',marker='o')
    plt.plot(yy1, xx1,  color='r',markerfacecolor='blue',marker='o')
    for a, b in zip(xx1, yy1):  
        plt.text(a, b, (a,b),ha='center', va='bottom', fontsize=10)  
    for a, b in zip(xx1, yy2):  
        plt.text(a, b, (a,b),ha='center', va='bottom', fontsize=10) 
    for a, b in zip(yy1, xx2):  
        plt.text(a, b, (a,b),ha='center', va='bottom', fontsize=10)  
    for a, b in zip(yy2, xx2):  
        plt.text(a, b, (a,b),ha='center', va='bottom', fontsize=10) 
    for a, b in zip(xx2, yy2):  
        plt.text(a, b, (a,b),ha='center', va='bottom', fontsize=10)  
    for a, b in zip(xx2, yy1):  
        plt.text(a, b, (a,b),ha='center', va='bottom', fontsize=10) 
    for a, b in zip(yy2, xx1):  
        plt.text(a, b, (a,b),ha='center', va='bottom', fontsize=10)  
    for a, b in zip(yy1, xx1):  
        plt.text(a, b, (a,b),ha='center', va='bottom', fontsize=10) 
    plt.legend()  
    plt.show()
        
bresenhamcircle(10)
```

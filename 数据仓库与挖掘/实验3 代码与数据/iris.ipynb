```python
import pandas as pd
from sklearn.utils import shuffle as reset
import matplotlib.pyplot as plt
import numpy as np
def train_test_split(data, test_size=0.2, shuffle=True, random_state=42):
    if shuffle:
        data = reset(data, random_state=random_state)

    train = data[int(len(data)*test_size):].reset_index(drop = True)
    test  = data[:int(len(data)*test_size)].reset_index(drop = True)
    
    return train, test


data = pd.read_csv('iris.csv')
class1 = data['species']
class1 = np.array(class1)
class1[class1=="Iris-setosa"]=0
class1[class1=="Iris-versicolor"]=1
class1[class1=="Iris-virginica"]=2
print(class1)
data = data.drop(['species'],axis=1)
# train_data,test_data = train_test_split(data)
# train_data.columns = ['sepal length', 'sepal width', 'petal length', 'petal width']
# test_data.columns = ['sepal length', 'sepal width', 'petal length', 'petal width']

data1=np.array(data)
plt.scatter(data1[:, 0], data1[:, 1], c=class1)
plt.show()
```

```python
from sklearn import cluster

k_means = cluster.KMeans(n_clusters=3, max_iter=50, random_state=1)
k_means.fit(data) 

all_predictions =k_means.predict(data)
plt.scatter(data1[:, 0], data1[:, 1], c=all_predictions)
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

numClusters = [1,2,3,4,5,6]
SSE = []
for k in numClusters:
    k_means = cluster.KMeans(n_clusters=k)
    k_means.fit(data)
    SSE.append(k_means.inertia_)

plt.plot(numClusters, SSE)
plt.xlabel('Number of Clusters')
plt.ylabel('SSE')
```

```python
import pandas as pd
import numpy as np
data = pd.read_csv('breast-cancer-wisconsin.data', delimiter=',')
data = np.array(data)
#data[data=='?'] = 0
data=np.delete(data, np.where(data=='?')[0], axis=0)
print(data)

plt.scatter(data[:, 0], data[:, 1], c=data[:, -1])
plt.show()
```

```python
from sklearn import cluster

k_means = cluster.KMeans(n_clusters=3, max_iter=50, random_state=1)
k_means.fit(data) 
all_predictions =k_means.predict(data)
plt.scatter(data[:, 0], data[:, 1], c=all_predictions)
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

numClusters = [1,2,3,4,5,6]
SSE = []
for k in numClusters:
    k_means = cluster.KMeans(n_clusters=k)
    k_means.fit(data)
    SSE.append(k_means.inertia_)

plt.plot(numClusters, SSE)
plt.xlabel('Number of Clusters')
plt.ylabel('SSE')
```

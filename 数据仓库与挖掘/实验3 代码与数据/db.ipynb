```python
X5, Y5 = datasets.make_classification(n_samples=200, n_features=2, n_redundant=0, n_informative=2, n_clusters_per_class=1)
plt.scatter(X5[:,0], X5[:,1], marker='o', c=Y5)
```

```python
from sklearn.cluster import DBSCAN

X1=pd.DataFrame(X5)
X1.columns=['x','y']
db = DBSCAN(eps=0.36, min_samples=5).fit(X1)
# appreciation = db.labels_
# plt.scatter(x=X1[:,0],y=X1[:,1],c=appreciation)
core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True
labels = pd.DataFrame(db.labels_,columns=['Cluster ID'])
result = pd.concat((X1,labels), axis=1)
result.plot.scatter(x='x',y='y',c='Cluster ID', colormap='jet')
```

```python
from sklearn import cluster
import pandas as pd

spectral = cluster.SpectralClustering(n_clusters=2,random_state=1,affinity='rbf',gamma=1)
spectral.fit(X1)
labels1 = pd.DataFrame(spectral.labels_,columns=['Cluster ID'])
result1 = pd.concat((X1,labels1), axis=1)
result1.plot.scatter(x='x',y='y',c='Cluster ID',colormap='jet')
```

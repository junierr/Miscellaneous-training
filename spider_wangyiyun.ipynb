```python
# -*- coding:utf-8 -*-
import sys
from bs4 import BeautifulSoup
import requests
import re
import os
import urllib
import time
import json
headers = {
    'User-Agent':'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.103 Safari/537.36'
}#伪装成浏览器
SingerId = input('请输入歌单ID:')
#SingerId = sys.argv[1]
print(SingerId)
#url = 'https://music.163.com/#/artist?id={}'.format(SingerId)
url = 'https://music.163.com/discover/toplist?id={}'.format(SingerId)
#url = 'https://music.163.com/discover/toplist?id=2250011882'
s = requests.session()
response = s.get(url, headers=headers).content
#print(response)
soup = BeautifulSoup(response, 'lxml')
data = soup.find('ul', class_= 'f-hide')
#print(data)
lists = []
for music in data.find_all('a'):
    #print('{}:{}'.format(music.text,music['href']))
    print(music)
    list = []
    music_url = 'http://music.163.com/song/media/outer/url' + music['href'][5:] + '.mp3'
    music_name = music.text
    list.append(music_name)
    list.append(music_url)
    lists.append(list)
    
if lists==[]:
    print('爬虫失败')
else:
    #print(lists)
    for i in lists:
        url = i[1]
        name = i[0]
        print(url, name)
#         try:
#             print('正在下载')
#             #urllib.request.urlretrieve(url, './网易云歌单/%s.mp3' % name)
#             val = os.system('wget -O "./网易云歌单/%s.mp3" %s' % (name, url))
#             print('下载成功')
#         except Exception as e:
#             print(e)
#             print('下载不成功')

#保存结果到json文件
#这里强制设定目标文件为utf-8，并且dump的时候设置不要ascii码，防止出现中文转义
with open('wangyiyun.json','w',encoding='utf-8') as fp:
    try:
        json.dump(lists,fp=fp,indent=4,ensure_ascii=False)
        print('保存成功')
    except Exception as e:
        print(e)
        print('保存失败')
```
